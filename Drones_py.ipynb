```python
import random
import time
import concurrent.futures
import unittest

# Clase Drone que representa cada drone
class Drone:
    def __init__(self, id):
        self.id = id
        self.x = random.randint(0, 100)
        self.y = random.randint(0, 100)
        self.z = random.randint(0, 100)
        self.colisiones_detectadas = []  # Lista para guardar colisiones detectadas

    def mover(self):
        # Simula el movimiento cambiando sus coordenadas
        self.x += random.randint(-5, 5)
        self.y += random.randint(-5, 5)
        self.z += random.randint(-5, 5)
        time.sleep(0.1)  # Simula el tiempo de movimiento
        print(f"Drone {self.id} se movió a ({self.x}, {self.y}, {self.z})")

    def verificar_colision(self, otros_drones):
        # Verifica si hay colisiones entre drones
        for drone in otros_drones:
            if self != drone:
                # Asegurarse de que las posiciones sean iguales para detectar la colisión
                if self.x == drone.x and self.y == drone.y and self.z == drone.z:
                    colision_message = f"¡Colisión detectada entre Drone {self.id} y Drone {drone.id}!"
                    self.colisiones_detectadas.append(colision_message)
                    print(colision_message)

# Función para mover drones de manera concurrente
def mover_drones_concurrentes(drones):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Mueve los drones concurrentemente
        executor.map(lambda drone: drone.mover(), drones)

    # Verifica las colisiones después de mover los drones
    for drone in drones:
        drone.verificar_colision(drones)

# Función principal para ejecutar el movimiento y la verificación de drones
def main():
    drones = [Drone(i) for i in range(10)]  # Crear 10 drones
    mover_drones_concurrentes(drones)  # Mover y verificar colisiones entre drones
    return drones  # Devuelve los drones para su verificación en las pruebas

# Clase de pruebas unitarias para la clase Drone
class TestDrone(unittest.TestCase):

    def test_mover(self):
        # Verifica que el drone se mueva (sus coordenadas cambien)
        drone = Drone(1)
        x_before = drone.x
        y_before = drone.y
        z_before = drone.z
        drone.mover()  # Mueve el drone
        self.assertNotEqual((x_before, y_before, z_before), (drone.x, drone.y, drone.z))

    def test_colision(self):
        # Verifica que se detecta una colisión cuando dos drones tienen la misma posición
        drone1 = Drone(1)
        drone2 = Drone(2)

        # Hacemos que los drones estén en la misma posición forzando las coordenadas
        drone1.x = drone2.x = 50  # Fuerza la misma posición X
        drone1.y = drone2.y = 50  # Fuerza la misma posición Y
        drone1.z = drone2.z = 50  # Fuerza la misma posición Z

        drones = [drone1, drone2]

        # Mover drones y verificar colisiones
        mover_drones_concurrentes(drones)

        # Verificamos que la colisión fue detectada en el drone1
        self.assertIn("¡Colisión detectada entre Drone 1 y Drone 2!", drone1.colisiones_detectadas)
        self.assertIn("¡Colisión detectada entre Drone 1 y Drone 2!", drone2.colisiones_detectadas)

# Si el archivo se ejecuta directamente, llamar a la función principal
if __name__ == "__main__":
    # Ejecuta la función principal para simular el movimiento y las colisiones
    drones = main()

    # Ejecutar pruebas unitarias
    unittest.main(argv=[''], exit=False)
```

```python
# Importación de las bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para gráficos
sns.set(style="whitegrid")

# URL del archivo CSV con datos de emisiones de CO2
url = 'https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv'

# Cargar los datos desde la URL
df = pd.read_csv(url)

# Mostrar las primeras filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())

# Verificación de las columnas y tipos de datos
print("\nColumnas del DataFrame:")
print(df.columns)
print("\nTipos de datos:")
print(df.dtypes)

# -------------------------------------------
# 1. Preprocesamiento

# Verificar si hay filas duplicadas
print("\nNúmero de filas duplicadas:")
print(df.duplicated().sum())  # Número de filas duplicadas
df = df.drop_duplicates()  # Eliminar duplicados

# Convertir la columna 'co2' a tipo float, si no es numérica
df['co2'] = pd.to_numeric(df['co2'], errors='coerce')

# Manejo de valores nulos (puedes elegir una estrategia de imputación o eliminación)
df = df.dropna(subset=['co2'])  # Eliminar filas con valores nulos en 'co2'

# Verificar si aún hay nulos
print("\nValores nulos después del preprocesamiento:")
print(df.isnull().sum())

# -------------------------------------------
# 2. Ordenar y Filtrar Datos
# Filtrar los datos para obtener solo las filas del año más reciente
latest_year = df['year'].max()
df_filtered_latest = df[df['year'] == latest_year]

# Ordenar los datos por las emisiones de CO2 en el año más reciente en orden descendente
df_sorted = df_filtered_latest.sort_values(by='co2', ascending=False)

# Mostrar los primeros 10 registros después de ordenar y filtrar
print("\nPrimeros 10 registros después de filtrar y ordenar:")
print(df_sorted.head(10))

# -------------------------------------------
# 3. Estadísticas Descriptivas
# Estadísticas descriptivas de las emisiones de CO2
print("\nEstadísticas descriptivas de las emisiones de CO2:")
print(df_sorted['co2'].describe())

# -------------------------------------------
# 4. Visualización de Datos

# 1. Gráfico de barras de los 10 países con mayores emisiones de CO2 en el año más reciente
top_10_countries = df_sorted.head(10)

# Asegurarse de que 'country' y 'co2' existen y son las columnas correctas
print("\nNombres de las columnas en top_10_countries:")
print(top_10_countries.columns)

# Verificar si las columnas 'country' y 'co2' existen
if 'country' not in top_10_countries.columns or 'co2' not in top_10_countries.columns:
    # Renombramos las columnas si no coinciden
    top_10_countries = top_10_countries.rename(columns={'Country Name': 'country', 'co2': 'co2'})

# Graficar
plt.figure(figsize=(12, 6))
sns.barplot(x='co2', y='country', data=top_10_countries, palette='viridis')
plt.title(f'Top 10 Países con Mayores Emisiones de CO2 en {latest_year}')
plt.xlabel('Emisiones de CO2')
plt.ylabel('País')
plt.savefig('top_10_countries_co2.png')
plt.show()

# 2. Gráfico de líneas de las emisiones de CO2 a lo largo del tiempo para un país específico (ejemplo: 'United States')
df_us = df[df['country'] == 'United States']

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='co2', data=df_us, marker='o', color='b')
plt.title('Emisiones de CO2 de Estados Unidos a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Emisiones de CO2')
plt.grid(True)
plt.savefig('us_co2_trend.png')
plt.show()

# 3. Gráfico de dispersión de las emisiones de CO2 a lo largo del tiempo
plt.figure(figsize=(12, 6))
sns.scatterplot(x='year', y='co2', data=df, palette='Set2')
plt.title('Emisiones de CO2 a lo largo del tiempo')
plt.xlabel('Año')
plt.ylabel('Emisiones de CO2')
plt.savefig('co2_scatter.png')
plt.show()

# 4. Mapa de calor de correlación entre variables (si hay más columnas numéricas)
# Verificar si existen otras columnas numéricas en el DataFrame
numeric_columns = df.select_dtypes(include=[np.number]).columns
print("\nColumnas numéricas para correlación:")
print(numeric_columns)

# Crear un mapa de calor de correlación
corr = df[numeric_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor de Correlación entre Variables')
plt.savefig('correlation_heatmap_co2.png')
plt.show()

# -------------------------------------------
# 5. Exportar Resultados
# Exportar los datos filtrados y ordenados a un archivo CSV
df_sorted.to_csv('top_10_co2_emissions.csv', index=False)
print("\nDatos exportados a 'top_10_co2_emissions.csv'.")

# Exportar los gráficos generados como imágenes PNG
print("\nLos gráficos han sido exportados a las siguientes imágenes:")
print("1. top_10_countries_co2.png")
print("2. co2_scatter.png")
print("3. us_co2_trend.png")
print("4. correlation_heatmap_co2.png")


print(".-.- aquí comienza la nueva parte del ejercicio")

# Asumimos que hay una columna 'Continent' que define el continente de cada país
# Si no existe, se puede agregar manualmente o usar una aproximación
# Aquí lo haremos manualmente como ejemplo (puedes modificarlo según tus datos)

df['Continent'] = df['country'].apply(lambda x: 'America' if x in ['United States', 'Brazil', 'Canada'] else 'Others')  # Ejemplo simplificado

# Agrupar por continente y año, y calcular la media de las emisiones de CO2 por continente
df_continent = df.groupby(['Continent', 'year'])['co2'].mean().reset_index()
print("los resultados obtenidos son", df_continent)

# Graficar el crecimiento de las emisiones de CO2 por continente a lo largo del tiempo
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='co2', hue='Continent', data=df_continent)
plt.title("Emisiones de CO2 por Continente")
plt.xlabel("Año")
plt.ylabel("Emisiones Media de CO2")
plt.legend(title='Continente')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import matplotlib.pyplot as plt

# Crear un conjunto de documentos
documents = ['Me gusta el fútbol', 'Me gusta la música', 'El fútbol es genial', 'El fútbol es mi deporte favorito']

# Crear el vectorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Convertir los documentos en una matriz TF-IDF
X_tfidf = tfidf_vectorizer.fit_transform(documents)

# Convertir a DataFrame para mejor visualización
df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Sumar los TF-IDF de cada palabra
tfidf_sums = df_tfidf.sum(axis=0).sort_values(ascending=False)

# Graficar las palabras con mayor TF-IDF en un gráfico de barras
plt.figure(figsize=(10, 6))
tfidf_sums.head(10).plot(kind='bar', color='orange')
plt.title('Palabras con Mayor TF-IDF')
plt.xlabel('Palabras')
plt.ylabel('TF-IDF')
plt.xticks(rotation=45)
plt.show()

# Exportar los resultados a un archivo Excel
df_tfidf.to_excel('tfidf_palabras.xlsx', index=False)
print("Archivo Excel exportado con éxito.")
```
