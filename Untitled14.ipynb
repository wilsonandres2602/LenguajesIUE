```python
# Importar las libreras necesarias
import numpy as np  # Para manejar operaciones matemticas y matrices
import matplotlib.pyplot as plt  # Para generar grficos
from scipy.integrate import solve_ivp  # Para resolver ecuaciones diferenciales

# Definicin del modelo Lotka-Volterra (presa-depredador)
def lotka_volterra(t, z, alpha, beta, delta, gamma):
    """
    Modelo Lotka-Volterra (Depredador-Presa)
    dx/dt = x - xy
    dy/dt = xy - y

    Parmetros:
    - t: Tiempo
    - z: Estado del sistema [x, y] (presas, depredadores)
    - alpha: Tasa de nacimiento de la presa
    - beta: Tasa de depredacin
    - delta: Tasa de reproduccin del depredador
    - gamma: Tasa de muerte del depredador
    """
    x, y = z  # Extraer las variables x (presas) y y (depredadores)
    dxdt = alpha * x - beta * x * y  # Ecuacin diferencial para la poblacin de presas
    dydt = delta * x * y - gamma * y  # Ecuacin diferencial para la poblacin de depredadores
    return [dxdt, dydt]  # Retornar las derivadas

# Definicin del modelo SIR (Epidemia)
def sir_model(t, z, beta, gamma):
    """
    Modelo SIR (Epidemia)
    dS/dt = -SI
    dI/dt = SI - I
    dR/dt = I

    Parmetros:
    - t: Tiempo
    - z: Estado del sistema [S, I, R] (susceptibles, infectados, recuperados)
    - beta: Tasa de infeccin
    - gamma: Tasa de recuperacin
    """
    S, I, R = z  # Extraer las variables S (susceptibles), I (infectados), R (recuperados)
    dSdt = -beta * S * I  # Cambio en la poblacin susceptible
    dIdt = beta * S * I - gamma * I  # Cambio en la poblacin infectada
    dRdt = gamma * I  # Cambio en la poblacin recuperada
    return [dSdt, dIdt, dRdt]  # Retornar las derivadas

# Funcin para resolver los sistemas de ecuaciones diferenciales
def solve_system(model, params, y0, t_span, t_eval):
    """
    Resuelve un sistema de ecuaciones diferenciales ordinarias.

    Parmetros:
    - model: Funcin que define el sistema de ecuaciones
    - params: Parmetros del modelo
    - y0: Condiciones iniciales
    - t_span: Intervalo de tiempo (t_min, t_max)
    - t_eval: Puntos de tiempo donde evaluar la solucin

    Retorna:
    - Solucin del sistema
    """
    solution = solve_ivp(model, t_span, y0=y0, t_eval=t_eval, args=params, method='RK45')
    return solution  # Devuelve la solucin del sistema

# Funcin para graficar los resultados
def plot_solution(solution, labels, title):
    """
    Grafica la solucin de un sistema de ecuaciones diferenciales.

    Parmetros:
    - solution: Resultado de solve_ivp
    - labels: Etiquetas de las variables
    - title: Ttulo del grfico
    """
    plt.figure(figsize=(10, 5))  # Tamao de la figura
    for i in range(len(labels)):  # Iterar sobre las variables a graficar
        plt.plot(solution.t, solution.y[i], label=labels[i])  # Graficar cada variable
    plt.xlabel('Tiempo')  # Etiqueta del eje X
    plt.ylabel('Poblacin')  # Etiqueta del eje Y
    plt.title(title)  # Ttulo del grfico
    plt.legend()  # Agregar leyenda
    plt.grid()  # Mostrar cuadrcula
    plt.show()  # Mostrar grfico

# ------------------ Modelo Lotka-Volterra ------------------

# Definir los parmetros para el modelo Lotka-Volterra
alpha, beta, delta, gamma = 0.1, 0.02, 0.01, 0.1  # Coeficientes del sistema
lv_params = (alpha, beta, delta, gamma)  # Agruparlos en una tupla
lv_y0 = [40, 9]  # Condiciones iniciales: 40 presas y 9 depredadores
t_span = (0, 200)  # Tiempo de simulacin de 0 a 200
t_eval = np.linspace(0, 200, 1000)  # Puntos en los que se evaluar la solucin

# Resolver el modelo Lotka-Volterra
lv_solution = solve_system(lotka_volterra, lv_params, lv_y0, t_span, t_eval)

# Graficar los resultados del modelo Lotka-Volterra
plot_solution(lv_solution, ['Presas', 'Depredadores'], 'Modelo Lotka-Volterra')

# ------------------ Modelo SIR ------------------

# Definir los parmetros para el modelo SIR
beta, gamma = 0.3, 0.1  # Tasa de infeccin y recuperacin
sir_params = (beta, gamma)  # Agruparlos en una tupla
sir_y0 = [990, 10, 0]  # Condiciones iniciales: 990 susceptibles, 10 infectados, 0 recuperados
t_span = (0, 100)  # Tiempo de simulacin de 0 a 100
t_eval = np.linspace(0, 100, 1000)  # Puntos en los que se evaluar la solucin

# Resolver el modelo SIR
sir_solution = solve_system(sir_model, sir_params, sir_y0, t_span, t_eval)

# Graficar los resultados del modelo SIR
plot_solution(sir_solution, ['Susceptibles', 'Infectados', 'Recuperados'], 'Modelo SIR')

# ------------------ Pruebas Unitarias ------------------

def test_lotka_volterra():
    """
    Prueba unitaria para el modelo Lotka-Volterra.
    Verifica que la solucin se haya calculado correctamente.
    """
    sol = solve_system(lotka_volterra, lv_params, lv_y0, t_span, t_eval)
    assert sol.success, "Error en la solucin de Lotka-Volterra"

def test_sir_model():
    """
    Prueba unitaria para el modelo SIR.
    Verifica que la solucin se haya calculado correctamente.
    """
    sol = solve_system(sir_model, sir_params, sir_y0, t_span, t_eval)
    assert sol.success, "Error en la solucin del modelo SIR"

# Ejecutar las pruebas unitarias
test_lotka_volterra()
test_sir_model()

# Mensaje de confirmacin si las pruebas pasan correctamente
print("Pruebas pasadas correctamente.")
print("Resultados del modelo Lotka-Volterra en t=0:", lv_solution.y[:, 0])
print("Resultados del modelo SIR en t=0:", sir_solution.y[:, 0])

```
